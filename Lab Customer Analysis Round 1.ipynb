```python
#estandarizamos las cabeceras de las columnas
file1.columns = file1.columns.str.lower().str.replace(' ', '_')#con esto quitamos las mayusculas y remplazmos espacios con _
```

```python
#concatenar los tres DF
combined_df = pd.concat([file1, file2, file3], ignore_index=True)
```

```python
combined_df.shape
```

```python
combined_df
```

```python
#Que columna son numericas
numerical_columns = combined_df.select_dtypes(include=['float64']).columns.tolist()
print("Numerical Columns:", numerical_columns)
```

```python
#que columnas son categorical
categorical_columns = combined_df.select_dtypes(exclude=['float64']).columns.tolist()
print("Categorical Columns:", categorical_columns)
```

```python
#borrar la columna education y the number of open complaints del DF
combined_df.drop(columns=['education', 'number_of_open_complaints'], inplace=True)
```

```python
combined_df
```

```python
#corrige los valores en la columna customer 
combined_df['customer_lifetime_value'] = pd.to_numeric(combined_df['customer_lifetime_value'].str.replace('%', '')) * 100
```

```python
#vamos a chequear si hay filas repetidas
combined_df.drop_duplicates(inplace=True)
```

```python
#eliminamos los clientes que tengan un ingreso de 0 o menos
combined_df = combined_df[combined_df['income'] > 0]
```
